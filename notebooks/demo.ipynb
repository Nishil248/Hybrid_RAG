```python
"""
notebooks/demo.py
A small script-style demo for interactive exploration (use instead of .ipynb).
"""

from pathlib import Path
import json
from src.hybrid_retrieval import HybridRetrievalSystem
import yaml

def load_config(path="config/default_config.yaml"):
    with open(path, "r") as f:
        return yaml.safe_load(f)

def main():
    cfg = load_config()
    system = HybridRetrievalSystem(cfg)
    # If you have prebuilt models:
    # system.vector_db.load("models/vector_db")
    # system.kg.load("data/knowledge_graphs/default_kg")
    # For demo, index tiny docs
    docs = [
        {"id": "d1", "title": "NLP", "text": "Natural language processing and deep learning"},
        {"id": "d2", "title": "Graphs", "text": "Knowledge graphs and graph neural networks"}
    ]
    system.index_documents(docs)
    q = "deep learning for language"
    print("Query:", q)
    out = system.retrieve(q, k=5)
    print("Results:", out)

if __name__ == "__main__":
    main()
```
